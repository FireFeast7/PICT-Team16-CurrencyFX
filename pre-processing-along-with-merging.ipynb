```python
file_paths = ['/kaggle/working/2012.csv', '/kaggle/working/2013.csv', '/kaggle/working/2014.csv', '/kaggle/working/2015.csv',
              '/kaggle/working/2016.csv', '/kaggle/working/2017.csv', '/kaggle/working/2018.csv', '/kaggle/working/2019.csv',
              '/kaggle/working/2020.csv', '/kaggle/working/2021.csv', '/kaggle/working/2022.csv']

dfs = []
for file_path in file_paths:
    df = pd.read_csv(file_path)
    dfs.append(df)
merged_df = pd.concat(dfs, axis=0, ignore_index=True)


```

```python
merged_df=merged_df.fillna(-1)
```

```python
merged_df=merged_df.drop(['Unnamed: 0'],axis=1)
```

```python
merged_df.to_csv("Merged_data_after_processing.csv")
```
